```python
import numpy as np
```

```python
nesm3_1985 = total_monthly_precip.sel(time='1985').mean(dim='time')
era5_1985 = tot_precip_mm.sel(valid_time='1985').mean(dim='valid_time')
```

```python
fig, ax = plt.subplots(figsize=(14, 7), subplot_kw=dict(projection=ccrs.PlateCarree()))

# Plot NESM3 in blue colormap
nesm3_plot = nesm3_1985.plot(
    ax=ax,
    cmap='Blues',
    alpha=0.7,
    add_colorbar=False,
    transform=ccrs.PlateCarree(),
    zorder=1
)

# Plot ERA5 in red colormap, overlay with transparency
era5_plot = era5_1985.plot(
    ax=ax,
    cmap='Reds',
    alpha=0.5,
    add_colorbar=False,
    transform=ccrs.PlateCarree(),
    zorder=2
)

# Add coastlines and borders for geographic context
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

# Add colorbars separately for clarity
cbar_nesm3 = plt.colorbar(nesm3_plot, ax=ax, orientation='vertical', fraction=0.046, pad=0.04)
cbar_nesm3.set_label('NESM3 Precipitation (mm/month)')

cbar_era5 = plt.colorbar(era5_plot, ax=ax, orientation='vertical', fraction=0.046, pad=0.12)
cbar_era5.set_label('ERA5 Precipitation (mm/month)')

plt.title('Average Monthly Precipitation in 1985\nNESM3 (blue) vs ERA5 (red)')
plt.show()
```

```python
nesm3_1985 = total_monthly_precip.sel(time='1985').mean(dim='time')
era5_1985 = tot_precip_mm.sel(valid_time='1985').mean(dim='valid_time')

fig, ax = plt.subplots(figsize=(14, 7), dpi=150, subplot_kw=dict(projection=ccrs.PlateCarree()))

levels = np.linspace(0, max(nesm3_1985.max().item(), era5_1985.max().item()), 50)

cs1 = ax.contourf(nesm3_1985.lon, nesm3_1985.lat, nesm3_1985, levels=levels, cmap='Blues', alpha=0.7, transform=ccrs.PlateCarree())
cs2 = ax.contourf(era5_1985.longitude, era5_1985.latitude, era5_1985, levels=levels, cmap='Reds', alpha=0.5, transform=ccrs.PlateCarree())

ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS.with_scale('10m'), linewidth=0.5)
ax.gridlines(draw_labels=True)

plt.colorbar(cs1, ax=ax, orientation='vertical', fraction=0.046, pad=0.04, label='NESM3 (mm/month)')
plt.colorbar(cs2, ax=ax, orientation='vertical', fraction=0.046, pad=0.12, label='ERA5 (mm/month)')

plt.title('Smoothed Average Monthly Precipitation in 1985\nNESM3 (blue) vs ERA5 (red)')
plt.show()
```
